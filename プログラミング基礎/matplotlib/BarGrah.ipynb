```python
labels = ['A','B','C','D','E']
man = [35,45,25,15,20]
woman = [20,25,30,40,20]
#alignのedgeは横にずらす
plt.bar(labels, man ,width=0.4,align='edge',color='b')
plt.bar(labels, woman ,width=0.4,align='center',color='coral')
plt.show()
```

```python
labels = ['A','B','C','D','E']
man = [35,45,25,15,20]
woman = [20,25,30,40,20]
plt.bar(labels, man ,width=0.4,align='edge',color='b',label='Man')
plt.bar(labels, woman ,width=0.4,align='center',color='coral',label='Woman')
plt.legend()
plt.show()
```

```python
plt.bar(labels, man ,width=0.4,align='edge',color='b',label='Man')
plt.bar(labels, woman ,width=0.4,align='center',color='coral',label='Woman')
plt.xlabel('Category')
plt.ylabel('Number of People')
plt.legend()
plt.title('Man and Woman')
plt.show()
```

```python
plt.bar(labels, man ,width=0.4,color='b',label='Man')
plt.bar(labels, woman ,width=0.4,color='coral',label='Woman',bottom=man)
plt.xlabel('Category')
plt.ylabel('Number of People')
plt.legend()
plt.title('Man and Woman')
plt.show()
```

```python
plt.barh(labels, man ,height=0.4,color='b',label='Man')
plt.barh(labels, woman ,height=0.4,color='coral',label='Woman',left=man)
plt.xlabel('Category')
plt.ylabel('Number of People')
plt.legend()
plt.title('Man and Woman')
plt.savefig('sample1.png')
plt.show()
```
