```python
import numpy as np
# 多次元配列も計算できる
```

```python
arr_x = np.array([1,2,3,4,5])
#  array配列にすることで、配列に対して一度に四則演算できる
arr_x/2
```

```python
x  = [1,2,3,4,5]
y =np.array([1,4,9,16,25])
# x,y　の個数を同じにしないといけない
plt.plot(x,y)
plt.plot(x,y/2)
plt.plot(x,y/3)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
plt.show()
```

```python
x  = [1,2,3,4,5]
y =np.array([1,4,9,16,25])
# x,y　の個数を同じにしないといけない
plt.plot(x,y, color = 'r')
plt.plot(x,y/2, color='b')
plt.plot(x,y/3, color='k')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
plt.show()
```

```python
x  = [1,2,3,4,5]
y =np.array([1,4,9,16,25])
plt.plot(x,y, color = 'red',label='A')
plt.plot(x,y/2, color='b',label='B')
plt.plot(x,y/3, color='k',label='C')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
# ラベルを表示させる
plt.legend()
plt.show()
```

```python
x  = [1,2,3,4,5]
y =np.array([1,4,9,16,25])
plt.plot(x,y, color = 'red',label='A')
plt.plot(x,y/2, color='b',label='B')
plt.plot(x,y/3, color='k',label='C')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
plt.legend(fontsize=15)
plt.show()
```
