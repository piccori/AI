```python
import matplotlib.pyplot as plt
```

```python
volume  = [10,20,30,20,10]
labels = ['A','B','C','D','E']

plt.bar(labels,volume)
plt.show()
```

```python
plt.bar(labels,volume,width=0.5)
plt.show()
```

```python
plt.bar(labels,volume,width=1)
plt.show()
```

```python
#水平方向
plt.barh(labels,volume)
plt.show()
```

```python
plt.bar(labels, volume, color='k')
plt.show()
```

```python
plt.bar(labels, volume, color='k',edgecolor='b')
plt.show()
```

```python
plt.bar(labels, volume,width=0.5,color='r',edgecolor='b',linewidth=3)
plt.show()
```

```python
plt.bar(labels, volume, color='r',edgecolor='b')
plt.xlabel('Category')
plt.ylabel('Volume')
plt.show()
```
