```python
from pandas_datareader import data
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
start = '2019-11-01'
end = '2020-11-01'
#第一引数に取得したいティッカーシンボル
# 第二引数にデータを取得するソース名
# ここの例ではYahooの日経平均株価
df = data.DataReader('^N225','yahoo', start, end)
```

```python
x = df.index#上の表でインデックスが日付になっている
y = df['Adj Close']#終値

plt.plot(x,y)
```

```python
import matplotlib.dates as mdates

plt.figure(figsize=(30,10))
plt.plot(x,y)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Date',fontsize=30)
plt.ylabel('Price',fontsize=30)
plt.grid()
plt.title('日経平均株価2019-2020',fontsize=35)
plt.annotate('point',xy=(mdates.date2num(x[10]),y[10]),xytext=(mdates.date2num(x[10])+10,24000),fontsize=30,color='red',arrowprops=dict(color='black'))
```

```python
plt.figure(figsize=(30,10))
plt.plot(x,y)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Date',fontsize=30)
plt.ylabel('Price',fontsize=30)
plt.grid()
plt.title('日経平均株価2019-2020',fontsize=35)
plt.annotate('points',xy=(mdates.date2num(x[10]),y[10]),xytext=(mdates.date2num(x[10])+10,24000),fontsize=30,color='red',arrowprops=dict(color='black'))
plt.savefig('N225.png')
```

```python
#移動平均線追加（5日,２５日）ある一定期間の価格から平均値を計算し、折れ線グラフで表したもの
span05 = 5
span25 = 25
# pandas のrollingメソッドで求める
#window は間隔を意味する
df['sma05'] = y.rolling(window=span05).mean()
df['sma25'] = y.rolling(window=span25).mean()
```

```python
df.head(100)
```

```python
plt.figure(figsize=(30,10))
plt.plot(x,y)
plt.plot(x,df['sma05'],label='sma05')
plt.plot(x,df['sma25'],label='sma25')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Date',fontsize=30)
plt.ylabel('Price',fontsize=30)
plt.grid()
plt.legend()
plt.title('日経平均株価2019-2020',fontsize=35)
```

```python
plt.figure(figsize=(30,10))
plt.plot(x,y,label='close',color='r')
plt.plot(x,df['sma05'],label='sma05',color='b')
plt.plot(x,df['sma25'],label='sma25',color='k')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Date',fontsize=30)
plt.ylabel('Price',fontsize=30)
plt.grid()
plt.legend()
plt.title('日経平均株価2019-2020',fontsize=35)
```
